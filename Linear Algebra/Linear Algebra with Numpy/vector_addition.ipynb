```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vec1 = np.array([1,0])
vec2 = np.array([0,1])
```

```python
vec3 = vec1+vec2
vec4 = vec1-vec2
vec5 = 4*vec3 + vec2/2
```

```python
print(vec5)
```

```python
plt.plot([0,vec1[0]],[0,vec1[1]],'ro-',label='v1')
plt.plot([0,vec2[0]],[0,vec2[1]],'bo-',label='v2')
plt.plot([0,vec3[0]],[0,vec3[1]],'g*-',label='v3')
plt.plot([0,vec4[0]],[0,vec4[1]],'kp-',label='v4')
plt.plot([0,vec5[0]],[0,vec5[1]],'kp-',label='v5')
```
